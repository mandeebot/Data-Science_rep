## PART A  REGRESSION ANALYSIS
for Regression problems, the goal is to forecast/ predict a continous output, the key differences here are in the type of models we import,
as opposed to a binary or a multi binary output as with the case with classification cases

### Phase 1
- Preparing and Processing Data for Modelling
  > data preprocessing,
  > variable encoding,
  > data scaling and normalization

```python
data.head(2)
```

```python
#encoding categorical features into numerical ones
data = pd.get_dummies(data,columns=["genre","dayofweek","season"],dtype="int64",drop_first=True)
```

```python
data.head(2)
```

```python
#segment data into predictor vars and target vars
y = data.pop('views_content')
```

```python
#scaling dataset
#importing the standard scaler
from sklearn.preprocessing import StandardScaler

#instantiate scaler object
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(data),columns=data.columns)
```

```python
scaled_data.head(2)
```

### Phase 2
- Splitting data into training and evaluation datasets
- Implementing Machine Learning/model Building and Training
- creating a predictions file

```python
#splitting data into training and validtion data sets
from sklearn.model_selection import train_test_split

X_train,x_test,y_train,y_test = train_test_split(scaled_data,y,train_size=0.8,random_state=1)
```

### building Base Model
- ols
- decision tree
- random forest regressor
- xgboost

```python
#import OLS(Linear regression models) from sckitlearn
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#import xgboost (assignment implement xgboost)
```

```python
#instantiating model
lin_reg = LinearRegression()

#fitting model on trainign data set
lin_reg.fit(X_train,y_train)
```

```python
lin_pred = lin_reg.predict(x_test)
#lin_pred
```

```python
#instantiating decision tree algo
dec_mod = DecisionTreeRegressor()
dec_mod.fit(X_train,y_train)
```

```python
dec_pred = dec_mod.predict(x_test)
#dec_pred
```

### Phase 3
- **we will compute the model's r_squared score (r2_score) which is a measure of how "good afit" the linear model is for modelling this kind of data**
- **here we will use business metrics such as the mean squared error(mse) and the root mean squared error (rsme) model Evaluation
to evaluate the linear regression's model's performance**
- we'ill also visualize the plot of the model's predicted label  vs the actual label

> FOR THE OLS/ Linear Regression Model

```python
#import business metrics from scklearn
from sklearn.metrics import mean_squared_error, r2_score

#compute the mean squared error score, RSME score and model's r-Squared value
mse = mean_squared_error(y_test,lin_pred)
r2 = r2_score(y_test,lin_pred)
rmse = np.sqrt(mse)

#print out metrics scores
print(f"Linear Regression's model(OLS) R_Squared Score: {r2}")

print(f"Linear Regression's model(OLS) Mean Squared Error Score: {mse}")
print(f"Linear Regression's model(OLS) RSME Score: {rmse}")
```

```python
#visualize model residuals
plt.scatter(y_test,lin_pred)
plt.xlabel("actual labels")
plt.ylabel("predicted labels")
plt.title(" model residuals")

z = np.polyfit(y_test,lin_pred,1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test),color='magenta')
plt.show()
```

> For the Decison tree algo, we will use the same business metrics introduced above, another method is we can also visualize tree nodes (research that area)

```python
#for decision tree algo
#compute the mean squared error score, RSME score and model's r-Squared value
dec_mse = mean_squared_error(y_test,dec_pred)
dec_r2 = r2_score(y_test,dec_pred)
dec_rmse = np.sqrt(dec_mse)

#print out metrics scores
print(f"Decision Tree model R_Squared Score: {dec_r2}")

print(f"Decision Tree model Mean Squared Error Score: {dec_mse}")
print(f"Decision Tree model RSME Score: {dec_rmse}")
```

```python
#visualize decision tree model residuals
plt.scatter(y_test,dec_pred)
plt.xlabel("actual labels")
plt.ylabel("predicted labels")
plt.title(" model residuals")

z = np.polyfit(y_test,dec_pred,1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test),color='magenta')
plt.show()
```

### Phase 4 Hyper Parameter 0ptimization
- hyperparameter optimization is the art and science of improving our model's performances
- we will be  implementing a and optimizing our decision tree Regressor model using cross-validation and RandomSearch technique to see if we can improve the Decision tree's performance on the data
- read more on hyperparameter optimization here: [Link](https://docs.google.com/document/d/1Na9x6LVqWmh2SvhXDGALJ0U4y2geQhgcVlcQX5-iQW4/edit?usp=sharing) 

```python
from sklearn.model_selection import KFold, cross_val_score, RandomizedSearchCV

#Set up kfold cross-validation
cv = KFold(n_splits=5, random_state=42, shuffle=True)

# set up the grid to modify hyper parameters from
grid={
    "ccp_alpha": [0,0.1,1],
    "max_features": [7, 10, 15],
    "min_samples_leaf": [1, 5, 20, 100],
    "min_samples_split": [2, 10, 50, 250],
    "criterion": ["friedman_mse","absolute_error","squared_error"],
    "max_depth": [4, 6, 8, 10, None]
}

# Find the best hyperparameter combination to optimize the R2 metric
dec_mod_search=RandomizedSearchCV(
    estimator=dec_mod,
    param_distributions=grid,
    n_iter=200,
    refit=True,
    return_train_score=True,
    cv=cv,    
    verbose=10,
    n_jobs=-1,
    random_state=42
)
```

```python
#Here we fit our "optimized model" on the training data set
%time tuned_model = dec_mod_search.fit(X_train, y_train)
```

```python
# we can print out and obseverve the best paramters for the model by using the 
tuned_model.best_params_
```

```python
#creating a predictions file
tuned_pred = tuned_model.predict(x_test)
```

```python
#evalutating performances of optimized model 
tuned_dec_mse = mean_squared_error(y_test,tuned_pred)
tuned_dec_r2 = r2_score(y_test,tuned_pred)
tuned_dec_rmse = np.sqrt(tuned_dec_mse)

print(f" R_Squared Score for optimized model is: {tuned_dec_r2}")

print(f"Mean Squared Error for optimized model is: {tuned_dec_mse}")
print(f" RSME Score for optimized model is: {tuned_dec_rmse}")
```

```python
#Residuals plot
#Plot predicted vs actual
plt.scatter(y_test,tuned_pred)
plt.xlabel("actual labels")
plt.ylabel("predicted labels")
plt.title(" model residuals")

# overlay the regression line
z = np.polyfit(y_test,tuned_pred,1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test),color='magenta')
plt.show()
```

```python
#comparing model metrics and selecting best performing model
model = {
    "Linear Regression":[r2,mse,rmse],
    "Decision Tree":[dec_r2,dec_mse,dec_rmse],
    "Optimized Model":[tuned_dec_r2,tuned_dec_mse,tuned_dec_rmse]
}
index=["R_Squared Score","Mean_Squared_Error_Score","Root_Mean_Squared_Error_Score"]
pd.DataFrame(model,index)
```

### Phase 3
- model evaluation and interpretation phase
- metrics to use for classification problems;
  >confusion matrix,
  >classification report

```python
#import metrics to use from sklearn
from sklearn.metrics import classification_report, confusion_matrix

#a helpful functin that helps us visualize the connfusion matrix
def confusion_matrix_sklearn(model, predictors, target):
    """
    To plot the confusion_matrix with percentages

    model: classifier
    predictors: independent variables
    target: dependent variable
    """
    y_pred = model.predict(predictors)
    cm = confusion_matrix(target, y_pred)
    labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
            for item in cm.flatten()
        ]
    ).reshape(2, 2)

    plt.figure(figsize=(6, 4))
    sb.heatmap(cm, annot=labels, fmt="")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
```
