# PYTHON LOOPS

### ● For Loop
A for loop acts as an iterator in Python; it goes through items that are in a sequence or any other
iterable item or performs a set of actions a specified number of times.. Objects that we've learned about that we can iterate over include strings, lists, tuples, and even built-in iterables for dictionaries, such as keys or values.

### ● While Loop
The while statement in Python is one of most general ways to perform iteration. A while statement will repeatedly execute a single statement or group of statements as long as the condition is true. The reason it is called a 'loop' is because the code statements are looped through over and over again until the condition is no longer met.


## FOR LOOP

Goes over every item in the sequence and runs all the code

The general format of a for loop is:

    for test:
        code statements
    

> For loops are a programming construct that let you go through each item in a sequence and then perform some operation on each one. For instance, you could use a for loop to go through all the values in a list, tuple, dictionary or series and check whether each conforms to some logical expression

## WHILE LOOP

 While loops are similar to for loops in that they allow you to execute code over and over again. For loops execute their contents, at most, a number of iterations equal to the length of the sequence you are looping over. While loops, on the other hand, keep executing their contents as long as a logical expression you supply remains true



### - Executes the loop continously until a condition is satisfied; iterates over a block of code as long as the condition is true.
> Ensure that for every While Loop, you have a condition that evaluates to False


The general format of a while loop is:

    while test:
        code statements

> While loops can cause trouble because they keep doing something until a condition you've set becomes false. If you set a condition that never becomes false and forget to add a way to stop the loop, your program will keep running forever, like a car stuck in traffic.

For example, if you have a loop that needs a certain number to increase each time, but you forget to make it increase, the loop will never stop because the condition will always be true. This can make your program freeze or crash.

## BREAK, CONTINUE, PASS

- Break: Forces out of a loop. Used to terminate(break out of) a loop entirely
- Continue: Ignores a particular iteration result. Skips to the next iteration of the loop
- Pass : Does Nothing; acts as a placeholder.
